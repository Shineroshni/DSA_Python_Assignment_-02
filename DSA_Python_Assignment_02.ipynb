```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# upload iris dataset
iris_data = pd.read_excel("iris-1.xlsx")
```

```python
# 1. Read the dataset into python environment.
iris_data
```

```python
# 2. Display the columns in the dataset.
print(iris_data.columns)
```

```python
# 3. Calculate the mean of each column in the dataset
mean_iris = iris_data[['SL','SW','PL','PW']].mean()
```

```python
print(mean_iris)
```

```python
# 4. Check for null values in the dataset.
null_iris = iris_data.isnull().sum()
```

```python
print(null_iris)
```

```python
# 5. Perform meaningful visualizations using the dataset.
```

```python
# Bar Chart
Barchart_iris = iris_data[['SL','SW','PL','PW']].mean()
Barchart_iris.plot(kind='bar',
                   color='green',
                   edgecolor='black')
plt.title('Bar Chart of Mean Values')
plt.xlabel('Feature')
plt.ylabel('mean')
plt.show()
```

```python
# Histogram
Histogram_iris = iris_data['SL'].hist(bins=20,
                                      color='red',
                                      edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('values')
plt.show()
```

```python
# Box plot
Boxplot_iris = sns.boxplot(x='SL',
                           y='Classification',
                           data=iris_data)
Boxplot_iris.set_title('Boxplot of Sepal Length by Classification')
plt.show()
```

```python
# Scatterplot
Scatterplot_iris = sns.scatterplot(x='PL',
                                   y='PW',
                                   data = iris_data,
                                   hue='Classification',
                                   palette='Set1')
Scatterplot_iris.set_title('Scatterplot of Petal Length vs Petal Width')
plt.xlabel('PL')
plt.ylabel('PW')
plt.show()
```
